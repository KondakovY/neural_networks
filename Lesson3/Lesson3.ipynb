# Датасет с винами

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_wine.html#sklearn.datasets.load_wine

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from keras.utils.vis_utils import plot_model
```

```python
data = load_wine()
```

```python
list(data.target_names)
```

```python
data.feature_names
```

```python
data.data.shape, data.target.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.25, shuffle=True, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train
```

```python
X_test
```

Нормализация данных:

```python
def norm(x):
  min = x.min(axis=0)
  max = x.max(axis=0)
  return (x - min) / (max - min)
```

```python
X_train = norm(X_train)
X_train
```

```python
X_test = norm(X_test)
X_test
```

```python
type(X_train)
```

```python
y_train
```

```python
y_test
```

Создание модели, подбор параметров:

```python
model = Sequential()
model.add(Dense(64, input_dim=13, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
%%time
model.fit(X_train, 
          to_categorical(y_train), 
          epochs=30, 
          batch_size=32)
```

Оценка модели на тестовой выборке:

```python
model.evaluate(
  X_test,
  to_categorical(y_test)
)
```

Вывод: 

- На фоне небольшого массива данных, модель показала хороший результат;

- На тестовйой выборке модель также показала хороший результат;

- Переобучение модели отсутствует.
